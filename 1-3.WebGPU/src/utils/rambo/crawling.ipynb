```python
import requests
from lxml import etree

CSSDESIGN_URL = "https://www.cssdesignawards.com/website-gallery?feature=WebGL&page={}"
urls = []
abouts = []
tags = []
categories = []
scores = []
dates = []

for i in range(108, 109):
    print("crawling: {}".format(i))
    text = requests.get(CSSDESIGN_URL.format(i), headers = {
        "User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36'
    }).text
    html = etree.HTML(text)
    if html is None:
        print("html is None in {}".format(i))
    # xpath = html.xpath('//*[@id="content"]/div[2]/div/div[1]/li[1]/div/figure/a')
    for j in range(1, 19):
        xpath = html.xpath('//*[@id="page"]/section/div/div[1]/article[{}]/div/div[1]/div/a[1]/@href'.format(j))
        if len(xpath) == 0:
            break
        cssdesign_out = "https://www.cssdesignawards.com/sites{}".format(xpath[0])
        ansText = requests.get(cssdesign_out, headers = {
            "User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36'
        }).text
        ansHtml = etree.HTML(ansText)
        ansUrl = ansHtml.xpath('//*[@id="page"]/section[1]/div/div/figure/a/@href')
        if len(ansUrl) == 0:
            break
        urls.append(ansUrl[0])
        tmp = ansHtml.xpath('//strong')
        hasAbout = False
        hasTag = False
        hasCategory = False
        hasScore = False
        for t in tmp:
            if t.text[0:5] == "ABOUT":
                abouts.append(t.tail)
                hasAbout = True
            if t.text[0:4] == "TAGS":
                parent = t.getparent()
                tags.append(parent.xpath('a/text()'))
                hasTags = True
            if t.text[0:8] == "CATEGORY":
                parent = t.getparent()
                categories.append(parent.xpath('a/text()')[0])
                hasCategories = True
        if not hasAbout:
            abouts.append("NONE")
        if not hasTags:
            tags.append([])
        if not hasCategories:
            categories.append([])
        tmp2 = ansHtml.xpath('//h5')
        for t in tmp2:
            if t.text[0:19] == "Final Judge's Score":
                parent = t.getparent()
                scores.append(parent.xpath('h3/text()')[0])
                hasScore = True
                break
        if not hasScore:
            scores.append("NONE")
        tmp3 = ansHtml.xpath('//h1/text()')
        if len(tmp3) == 0:
            dates.append("NONE")
        else:
            dates.append(tmp3[0])
dataFrame = pd.DataFrame({'url': urls, 'about': abouts, 'tags': tags, 'category': categories, 'score': scores, 'date': dates})
dataFrame.to_csv('cssdesign.csv', index=False)
```

```python
import requests
from lxml import etree
import pandas as pd
import queue

f = open("poki_urls.txt", "w")
url_template = "https://poki.com{}"
ans_url = set()
has_visit = set()
todo_url = queue.Queue()
todo_url.put(url_template.format(""))

while not todo_url.empty():
    url = todo_url.get()
    print(url)
    text = requests.get(url).text
    html = etree.HTML(text)
    sons = html.xpath('//a/@href')
    for son in sons:
        if son[0:5] == "/zh/g":
            if son not in has_visit:
                ans_url.add(url_template.format(son))
                has_visit.add(son)
        elif son[0:4] == "/zh/":
            if son not in has_visit:
                todo_url.put(url_template.format(son))
                has_visit.add(son)

for url in ans_url:
    f.write(url + "\n")
```
