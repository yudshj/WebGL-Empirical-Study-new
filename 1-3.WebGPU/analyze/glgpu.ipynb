```python
import pandas as pd
import json
from pathlib import Path
```

```python
f = Path('../output-{}/canvas-type'.format('windows')) / 'GALLERY-AWWWARDS-0022.json'
print(json.dumps(json.loads(f.read_text()), indent=2))
```

```python
lst = {}
oses = ['windows', 'linux']
for os in oses:
    p = Path('../output-{}/canvas-type'.format(os))
    for f in p.glob('*.json'):
        with f.open() as file:
            data = json.load(file)
            idx = data['idx']
            if idx not in lst:
                url = data['url']
                category = idx.split('-')[0]
                date = data['date']
                netIdleTimeout = data['netIdleTimeout']
                o = {}
                o['url'] = url
                o['idx'] = idx
                o['date'] = date
                o['category'] = category
                lst[idx] = o
                # for ooo in oses:
                #     o[f'{ooo}.netIdleTimeout'] = netIdleTimeout
                #     o[f'{ooo}.navigator.gpu'] = False
                #     o[f'{ooo}.bitmaprenderer'] = 0
                #     o[f'{ooo}.2d'] = 0
                #     o[f'{ooo}.webgl1'] = 0
                #     o[f'{ooo}.webgl2'] = 0
                #     o[f'{ooo}.webgl'] = 0
                #     o[f'{ooo}.webgpu'] = 0
            o = lst[idx]
            ooo = os
            o[f'{ooo}.netIdleTimeout'] = netIdleTimeout
            o[f'{ooo}.navigator.gpu'] = False
            o[f'{ooo}.bitmaprenderer'] = 0
            o[f'{ooo}.2d'] = 0
            o[f'{ooo}.webgl1'] = 0
            o[f'{ooo}.webgl2'] = 0
            o[f'{ooo}.webgl'] = 0
            o[f'{ooo}.webgpu'] = 0
            for canv in data['canvas']:
                if 'data' not in canv['value']:
                    continue
                d = canv['value']['data']
                if isinstance(d, dict):
                    o[f'{os}.navigator.gpu'] = d['navigator.gpu'] or o[f'{os}.navigator.gpu']
                    if 'bitmaprenderer' in d:
                        o[f'{os}.bitmaprenderer'] += d['bitmaprenderer']
                    if '2d' in d:
                        o[f'{os}.2d'] += d['2d']
                    if 'experimental-webgl' in d:
                        o[f'{os}.webgl1'] += d['experimental-webgl']
                        o[f'{os}.webgl'] += d['experimental-webgl']
                    if 'webgl' in d:
                        o[f'{os}.webgl1'] += d['webgl']
                        o[f'{os}.webgl'] += d['webgl']
                    if 'webgl2' in d:
                        o[f'{os}.webgl2'] += d['webgl2']
                        o[f'{os}.webgl'] += d['webgl2']
                    if 'webgpu' in d:
                        o[f'{os}.webgpu'] += d['webgpu']
print(len(lst))
df = pd.json_normalize(lst.values()).set_index('idx')
```
