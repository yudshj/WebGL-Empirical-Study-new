```python
import re

def trim(s: str):
    try:
        PATTERN = re.compile(r'// INITIAL HLSL BEGIN\n\n(.*)\n\n// INITIAL HLSL END', re.DOTALL)
        m = PATTERN.search(s)
        if m:
            return m.group(1)
        else:
            return s
    except:
        return s
```

```python
SUM_PREFIX = [
    'maghsk.counter.resources.',
    'maghsk.counter.vertexCount',
    'maghsk.counter.primitiveCount.',
    'maghsk.counter.bufferDataSent',
    'maghsk.counter.bufferSubDataSent',
    'maghsk.counter.funcCount.',
    'maghsk.pixelsRead',
    'canvasSize',
]

UNION_PREFIX = [
    'maghsk.extensions'
]
```

```python
sum_columns = sorted(set(c for c in df_b_rafs_new.columns if any([c.startswith(p) for p in SUM_PREFIX])))
union_columns = sorted(set(c for c in df_b_rafs_new.columns if any([c.startswith(p) for p in UNION_PREFIX])))
```

```python
def sum_texture_size(l: list):
    try:
        ret = 0
        for x in l:
            try:
                ret += x[1] * x[2]
            except:
                pass
        return ret
    except:
        print(l)
        return 0
```

```python
def get_func(key):
    def get_length(l):
        ret = 0
        if isinstance(l, list):
            for x in l:
                if isinstance(x[key], str):
                    ret += len(x[key])
        return ret
    return get_length
```

```python
import re

def func(str):
    begin_match = re.search(r'HSL BEGIN', str)
    end_match = re.search(r'HSL END', str)
    if begin_match and end_match:
        begin_pos = begin_match.end()
        end_pos = end_match.start()
        content = str[begin_pos:end_pos]
        return content
    else:
        return str
```

```python
def get_final(df):
    tmp = df[['index_frame', 'index_webapp', 'url'] + sum_columns].copy()
    tmp['canvasSize'] = df['canvasSize'].map(lambda x: x[0]*x[1])
    tmp['textureSize'] = df['maghsk.counter.textureFormat'].map(sum_texture_size) + df['maghsk.counter.compressedTextureFormat'].map(sum_texture_size)
    tmp['shaderLength'] = df['maghsk.shaders'].map(get_func('source'))
    tmp['shaderLengthTranslated'] = df['maghsk.shaders'].map(get_func('translatedSource'))
    ret = tmp.groupby(['index_webapp']).sum(numeric_only=True).drop(columns=['index_frame'])
    return ret
```

```python
df_a_raf_final = get_final(df_a_rafs_new)
df_b_raf_final = get_final(df_b_rafs_new)
```

```python
df_a_raf_final_reindex = df_a_raf_final.reindex(df_b_raf_final.index, fill_value=0)
df_a_raf_final_reindex
```

```python
df_ab_final = df_b_raf_final - df_a_raf_final_reindex
df_ab_final['canvasSize'] = df_b_raf_final['canvasSize']
df_ab_final['textureSize'] = df_b_raf_final['textureSize']
df_ab_final
```
