```python
import pandas as pd
```

```python
df = pd.read_excel('labeled_data/0328-all_url_list-0404.xlsx', index_col=0)
# rename columns
print(df.columns)
# df['comments'] = df['comments'].astype('string')
# df['category'] = df['category'].astype('string')
# df['url'] = df['url'].astype('string')
# df['ERROR'] = df['ERROR'].astype('Int64')
# df['usecase'] = df['usecase'].astype('Int64')
# df['interactions'] = df['interactions'].astype('Int64')
df = df.astype({
    'comments': 'string',
    'category': 'string',
    'url': 'string',
    'ERROR': 'Int64',
    'usecase': 'Int64',
    'interactions': 'Int64',
})
```

```python
df.comments = df.comments.fillna('')
df
```

```python
loc_error_na = (df['comments'].isna() & df['ERROR'].isna())
df.loc[loc_error_na, 'ERROR'] = 0
```

```python
loc_usecase_na = (df['ERROR'] == 0) & df['usecase'].isna()
df.loc[loc_usecase_na, 'usecase'] = 1
```

```python
loc_usecase_3_na = (df['ERROR'] == 0) & (df['usecase'] == 3) & df['interactions'].isna()
df.loc[loc_usecase_3_na, 'interactions'] = 0
```

```python
loc_error = (df['ERROR'] != 0) & (~df['usecase'].isna() | ~df['interactions'].isna())
df[loc_error]
```

```python
df.loc[loc_error, 'interactions'] = None
df.loc[loc_error, 'usecase'] = None
```

```python
loc_url_error = df.url.str[7:].str.contains('://')
tmp = df[loc_url_error]
tmp
```

```python
df.loc[loc_url_error, 'ERROR'] = 1
df.loc[loc_url_error, 'usecase'] = None
df.loc[loc_url_error, 'interactions'] = None
df.loc[loc_url_error, 'comments'] = 'URL error'
```

```python
df.loc[df.comments == 'File not found', 'ERROR'] = 1
df.loc[df.comments == '白屏', 'comments'] = 'white'
```

```python
df.to_pickle('output/0328-all_url_list.pkl')
df.to_excel('output/0328-all_url_list.xlsx')
```
