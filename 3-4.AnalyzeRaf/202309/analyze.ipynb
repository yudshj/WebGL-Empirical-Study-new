```python
import pandas as pd
```

```python
df = pd.read_pickle('df_raf-0916.pkl.zstd', compression='zstd')
```

```python
df
```

```python
df_ab = pd.json_normalize(df['frame'])
```

```python
df_b_rafs = df_ab['b_rafs'].explode().reset_index(drop=False).rename(columns={'index': 'index_webapp'})
df_b_rafs
```

```python
tmp = pd.json_normalize(df_b_rafs['b_rafs'].map(lambda x: x['value'])) # type: ignore
df_b_rafs_new = pd.concat([df_b_rafs['index_webapp'], tmp], axis=1)
df_b_rafs_new = df_b_rafs_new.explode('data.contextInfo').reset_index(drop=False).rename(columns={'index': 'index_frame'}).dropna(subset=['data.contextInfo']).reset_index(drop=True)
tmp = pd.json_normalize(df_b_rafs_new['data.contextInfo']) # type: ignore
df_b_rafs_new = pd.concat([df_b_rafs_new.drop(columns=['data.contextInfo']), tmp], axis=1)
```
