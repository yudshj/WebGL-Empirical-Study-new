```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

pickle_output_path = Path('output/pickle')
```

```python
# save to pickle first
df_url = pd.read_pickle(pickle_output_path / '0417-raf_url.pkl.zstd', compression='zstd')
df_context = pd.read_pickle(pickle_output_path / '0417-raf_context.pkl.zstd', compression='zstd')
df_program = pd.read_pickle(pickle_output_path / '0417-raf_program.pkl.zstd', compression='zstd')
df_raflist_abc = pd.read_pickle(pickle_output_path / '0417-raf_raflist_abc.pkl.zstd', compression='zstd')
df_raf_stat = pd.read_pickle(pickle_output_path / '0417-raf_stat.pkl.zstd', compression='zstd')
```

```python
df_raf_stat_subset = df_raf_stat[['rafList-semanticVertexCalled-POINTS', 'rafList-semanticVertexCalled-LINES', 'rafList-semanticVertexCalled-LINE_LOOP', 'rafList-semanticVertexCalled-LINE_STRIP', 'rafList-semanticVertexCalled-TRIANGLES', 'rafList-semanticVertexCalled-TRIANGLE_STRIP', 'rafList-semanticVertexCalled-TRIANGLE_FAN', 'rafList-semanticVertexCount-POINTS', 'rafList-semanticVertexCount-LINES', 'rafList-semanticVertexCount-LINE_LOOP', 'rafList-semanticVertexCount-LINE_STRIP', 'rafList-semanticVertexCount-TRIANGLES', 'rafList-semanticVertexCount-TRIANGLE_STRIP', 'rafList-semanticVertexCount-TRIANGLE_FAN']]
df_raf_stat_subset
```

```python
df_mean = df_raf_stat_subset[[x for x in df_raf_stat_subset.columns if x[1] == 'mean']]
df_mean.columns = [x[0][22:] for x in df_mean.columns]

df_count = df_raf_stat_subset[[x for x in df_raf_stat_subset.columns if x[1] == 'count']]
df_count.columns = [x[0][22:] for x in df_count.columns]
```

```python
df_mean = df_mean[df_mean.sum(axis=1) >= 1]
print(len(df_mean))
```

```python
df_mean
```

```python
to_plot = df_mean.describe(percentiles=[0.25, 0.5, 0.75, 0.9, 0.95, 0.99])

# sns plot with heatmap
plt.figure(figsize=(18, 18))
sns.heatmap(to_plot.T, annot=True, fmt='.2f', cmap='YlGnBu', vmin=0, vmax=1)
plt.title('df_mean RAF Statistics')
```

```python
to_plot = df_count.describe(percentiles=[0.25, 0.5, 0.75, 0.9, 0.95, 0.99])

# sns plot with heatmap
plt.figure(figsize=(18, 18))
sns.heatmap(to_plot.T, annot=True, fmt='.2f', cmap='YlGnBu', vmin=0, vmax=1)
plt.title('df_count RAF Statistics')
```

```python
col = 'Count-TRIANGLES'
palette = sns.color_palette('YlGnBu', 1)
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(df_count[col].sort_values().values, label=col, color=palette[0])
ax.set_title('RAF Statistics')
ax.set_xlabel('RAF')
ax.set_ylabel('Count')
ax.legend()
plt.show()
```

```python
col = 'Count-TRIANGLES'
palette = sns.color_palette('YlGnBu', 1)
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(df_mean[col].sort_values().values, label=col, color=palette[0])
ax.set_yscale('log')
ax.set_title('RAF Statistics')
ax.set_xlabel('RAF')
ax.set_ylabel('Count')
ax.legend()
plt.show()
```
