```python
tmp = df_simple[[
    'frame.data.contextInfo.memoryInfo.resources.framebuffer',
    'frame.data.contextInfo.memoryInfo.resources.renderbuffer',
    'frame.data.contextInfo.memoryInfo.resources.program',
    'frame.data.contextInfo.memoryInfo.resources.query',
    'frame.data.contextInfo.memoryInfo.resources.shader',
    'frame.data.contextInfo.memoryInfo.resources.sampler',
    'frame.data.contextInfo.memoryInfo.resources.texture',
    'frame.data.contextInfo.memoryInfo.resources.transformFeedback',
    'frame.data.contextInfo.memoryInfo.resources.vertexArray',
    'frame.data.contextInfo.memoryInfo.resources.sync'
]]

sns.displot(tmp / tmp.max(axis=0), kind='kde', height=5, aspect=2)
plt.xlim(0, 0.1)
```

```python
tmp = df_simple[[
    'frame.data.contextInfo.memoryInfo.memory.texture',
    'frame.data.contextInfo.memoryInfo.memory.buffer',
    'frame.data.contextInfo.memoryInfo.memory.renderbuffer',
    'frame.data.contextInfo.memoryInfo.memory.drawingbuffer',
    'frame.data.contextInfo.memoryInfo.memory.total',
 ]]

sns.displot(tmp / tmp.max(axis=0), kind='kde', height=5, aspect=2)
plt.xlim(0, 0.15)
```

```python
to_plot = [
    # 'frame.data.contextInfo.memoryInfo.resources.framebuffer',
    # 'frame.data.contextInfo.memoryInfo.resources.renderbuffer',
    'frame.data.contextInfo.memoryInfo.resources.program',
    'frame.data.contextInfo.memoryInfo.resources.shader',
    # 'frame.data.contextInfo.memoryInfo.resources.sampler',
    'frame.data.contextInfo.memoryInfo.resources.texture',
    # 'frame.data.contextInfo.memoryInfo.resources.transformFeedback',
    'frame.data.contextInfo.memoryInfo.resources.vertexArray',
]
```

```python
tmp = df_simple[to_plot].rename(columns={
    'frame.data.contextInfo.memoryInfo.resources.program': 'Program',
    'frame.data.contextInfo.memoryInfo.resources.shader': 'Shader',
    'frame.data.contextInfo.memoryInfo.resources.texture': 'Texture',
    'frame.data.contextInfo.memoryInfo.resources.vertexArray': 'Vertex Array',
})
tmp = tmp[tmp.sum(axis=1) > 0]
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 3))
for i, ax in enumerate(axes.flatten()):
    # sns.histplot(tmp.iloc[:, i][tmp.iloc[:, i] > 0], ax=ax, bins=20)
    taa = tmp.iloc[:, i]
    sns.histplot(taa[(taa > 0) & (taa < taa.quantile(0.99))], ax=ax, bins=50)
    # sns.scatterplot(x=taa.index, y=taa.values, ax=ax)
    # sns.boxplot(x=tmp.iloc[:, i], ax=ax)
    # ax.set_xscale('log')
    # ax.set_yscale('log')
    ax.set_ylabel('')
    # y scientific notation
    ax.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
    ax.set_title(tmp.columns[i])
axes[0].set_ylabel('Count')
plt.savefig('fig/results_context_resources.pdf', bbox_inches='tight')
    
```
